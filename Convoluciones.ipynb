```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
imagen = cv2.imread('iron.jpg', cv2.IMREAD_COLOR)
imagengris = cv2.imread('iron.jpg', cv2.IMREAD_GRAYSCALE)
imagenrgb = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB)
cv2.imwrite('iron.jpg', imagenrgb)

edge_kernel = np.array([[-1,-1,-1], [-1,8,-1], [-1,-1,-1]])
img1 = cv2.filter2D(imagen, -1, edge_kernel) 

gaussian_kernel = cv2.getGaussianKernel(60, 0)
img2 = cv2.filter2D(imagen, -1, gaussian_kernel)
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 2)
plt.title('Deteccion de bordes')
plt.imshow(img1)

plt.subplot(1, 3, 1)
plt.title('Imagen original')
plt.imshow(imagen)

plt.subplot(1, 3, 3)
plt.title('Desenfoque gaussiano')
plt.imshow(img2)

plt.tight_layout()
plt.show()
```
