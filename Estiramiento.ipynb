```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img = cv2.imread('iron.jpg',  cv2.IMREAD_GRAYSCALE)

hist_original = cv2.calcHist([img], [0], None, [256], [0, 256])


int_minima = np.min(img)
int_maxima = 200

img_est = np.uint8(((img - int_minima) / (int_maxima - int_minima)) * 255)

hist_est = cv2.calcHist([img_est], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.title('Imagen Original')
plt.imshow(img, cmap='gray')

plt.subplot(2, 2, 2)
plt.title('Histograma Original')
plt.plot(hist_original)
plt.xlim([0, 256])


plt.subplot(2, 2, 3)
plt.title('Imagen Estirada')
plt.imshow(img_est, cmap='gray')

plt.subplot(2, 2, 4)
plt.title('Histograma Estirado')
plt.plot(hist_est)
plt.xlim([0, 256])


plt.tight_layout()
plt.show()
```
