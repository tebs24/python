```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

imagen = cv2.imread('iron.jpg')
blue, green, red = cv2.split(imagen)

hist_original = cv2.calcHist([imagen], [0], None, [256], [0, 256])
bhist_original = cv2.calcHist([blue], [0], None, [256], [0, 256])
ghist_original = cv2.calcHist([green], [0], None, [256], [0, 256])
rhist_original = cv2.calcHist([red], [0], None, [256], [0, 256])

int_minima = np.min(imagen)
int_maxima = np.max(imagen)

desp_minimo=50
desp_maximo=200

img_est = np.uint8((imagen-int_minima)/(int_maxima-int_minima)*(desp_maximo-desp_minimo)+desp_minimo)
blue_e, green_e, red_e = cv2.split(img_est)

bhist_est = cv2.calcHist([blue_e], [0], None, [256], [0, 256])
ghist_est = cv2.calcHist([green_e], [0], None, [256], [0, 256])
rhist_est = cv2.calcHist([red_e], [0], None, [256], [0, 256])
hist_est = cv2.calcHist([img_est], [0], None, [256], [0, 256])

plt.figure(figsize=(12, 6))

plt.subplot(2, 3, 1)
plt.title('Imagen Original')
plt.imshow(imagen)

plt.subplot(2, 3, 2)
plt.title('Histograma Original')
plt.plot(hist_original)
plt.xlim([0, 256])

plt.subplot(2, 3, 3)
plt.title('Histograma por Canales Original')
plt.plot(bhist_original)
plt.plot(ghist_original, color='green')
plt.plot(rhist_original, color='red')

plt.subplot(2, 3, 4)
plt.title('Imagen Estirada')
plt.imshow(img_est)

plt.subplot(2, 3, 5)
plt.title('Histograma Estirado')
plt.plot(hist_est)
plt.xlim([0, 256])

plt.subplot(2, 3, 6)
plt.title('Histograma por Canales Estirado')
plt.plot(bhist_est)
plt.plot(ghist_est, color='green')
plt.plot(rhist_est, color='red')

plt.tight_layout()
plt.show()
```
